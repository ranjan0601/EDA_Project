```python
data = {
 'EmployeeID': [1, 2, 3, 4, 5],

 'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],

 'Department': ['HR', 'IT', 'IT', 'HR', 'Finance'],

 'Salary': [60000, 70000, 80000, 65000, 75000],

 'JoiningDate': ['2020-01-15', '2019-06-20', '2018-07-23', '2020-02-10', '2021-03-15'],

 'PerformanceScore': [3, 4, 2, 5, 3]
}
```

```python
# Creating a dataframe
import pandas as pd
df = pd.DataFrame(data)
df
```

```python
# Calculating the average salary of each department

avg_salary = df.groupby('Department')['Salary'].mean()
avg_salary
```

```python
# write a function to find employee with highest performance score in each department.
def highest_performance_by_department(df):
    # Find the employee with the highest performance score in each department
    idx = df.groupby('Department')['PerformanceScore'].idxmax()
    print(idx)
    highest_performers = df.loc[idx]
    return highest_performers

# Get the result
result = highest_performance_by_department(df)
print(result)
```

```python
# Add a column of number of year each employee worked with the compnay based on joining date

# Function to calculate the number of years and days
def calculate_years_days(joining_date, current_date):
    delta = current_date - joining_date
    years = delta.days // 365
    days = delta.days % 365
    return f"{years} years, {days} days"
```

```python
df['YearsWithCompany'] = df['JoiningDate'].dt.date.apply(lambda x: calculate_years_days(x, current_date))
df
```

```python
# 4. Create a pivot table to display the total salary and average performance score for each department.
pv2 = df.pivot_table(index= 'Department', values= ['Salary', 'PerformanceScore'], aggfunc= {'Salary': 'sum', 'PerformanceScore':'mean'})
pv2
```

```python
# How can you calculate the cumulative sum of the 'PerformanceScore' column grouped by 'Department'?


cumm_of_per_score = df.pivot_table(index='Department', values= 'PerformanceScore', aggfunc= 'cumsum')
cumm_of_per_score
```

```python
# Write a function to rank employees within each department based on their 'Salary'.

def rank_emp(df):
    df['Rank'] = df.groupby('Department')['Salary'].rank(method='dense', ascending=True)
    return df
```
